```python
import pandas as pd
x_train = pd.read_csv(r'C:\Users\brian\OneDrive\Desktop\JUNSOO\bigdata\data\bike_x_train.csv' , encoding = 'cp949') 
y_train = pd.read_csv(r'C:\Users\brian\OneDrive\Desktop\JUNSOO\bigdata\data\bike_y_train.csv' , encoding = 'cp949')
x_test = pd.read_csv(r'C:\Users\brian\OneDrive\Desktop\JUNSOO\bigdata\data\bike_x_test.csv' , encoding = 'cp949')

x_train.info()
```

```python
# 독립변수와 종속변수 관계 확인하기 (count = 자전거 대여량)
data = pd.concat([x_train , y_train] , axis = 1)        #세로방향 통합
print(data.head())

# 계절에 따른 대여량
print(data
      .groupby(['계절'])['count']
      .sum()
      )

# 공휴일에 따른 대여량
print(data
      .groupby(['공휴일'])['count']
      .sum()
      )
```

```python

# 근무일에 따른 대여량
print(data
      .groupby(['근무일'])['count']
      .sum()
      )

# 날씨에 따른 대여량
print(data
      .groupby(['날씨'])['count']
      .sum())
```

```python
### 데이터 전처리하기

## 파생변수만들기

#1)
# datatime 칼럼의 데이터 타입을 날짜 타입(datatime)으로 변경하기   ## 추후 test 데이터에 대해서도 동일하게 시행!
x_train['datetime'] = pd.to_datetime(x_train['datetime'])       # datatime 은 연/월/요일/시간대 값을 가짐
                                                                # 각각을 출력하여 요일 데이터를 만들기

# x_train의 'datatime' 칼럼에서 연도 데이터를 추출하여 'year' 칼럼에 저장하기
x_train['year'] = x_train['datetime'].dt.year       #dt 문법..?****

# x_train의 'year' 칼럼에서 중복 제거한 값을 확인하기
print(x_train['year'].unique())
```

```python
#2)
# x_train의 'datatime' 칼럼에서 월 데이터를 추출하여 'month' 칼럼에 저장하기
# datafram['str'] ==> 새로 정의한 변수// 이 형태로 데이터 추출하면 새로운 칼럼으로 저장됨! ***
x_train['month'] = x_train['datetime'].dt.month

print(x_train['month'].unique())

# x_train의 'datetime' 칼럼에서 시간 데이터를 추출
x_train['hour'] = x_train['datetime'].dt.hour
print(x_train['hour'].unique())

# x_train의 'datatime' 칼럼에서 week 데이터를 추출
x_train['dayofweek'] = x_train['datetime'].dt.dayofweek
print(x_train['dayofweek']
      .unique())
```

```python
# 파생변수가 포함된 독립변수와 종속변수를 통합한 data2 만들기
# x_train.info()      # 새로운 칼럼이 저장됐다.***

data2 = pd.concat([x_train , y_train] , axis = 1)

print(data2.groupby(['year'])['count'].sum())       #연도별로 변수간의 관계를 확인할 수 있게 됐다.
                                                    #연도별 차이가 유의미 하므로 분석에 포함

print(data2.groupby(['month'])['count'].sum())      #월별 차이가 무의미 하므로 분석에서 제외
                                                    #일별, 요일도 마찬가지

data2.head()
```

```python
# 분석에 포함시킬 데이터 칼럼

print(data2.groupby(['hour'])['count'].sum())       #시간별 차이가 유의미 하므로 분석에 포함

print(data2.groupby(['year'])['count'].sum())       #연도별로 변수간의 관계를 확인할 수 있게 됐다.
                                                    #연도별 차이가 유의미 하므로 분석에 포함

# test 데이터에 대해서도 동일하게 칼럼을 추가 *** 
x_test['datetime'] = pd.to_datetime(x_test['datetime'])
x_test['year'] = x_test['datetime'].dt.year
x_test['hour'] = x_test['datetime'].dt.hour


```

```python
# 불필요한 칼럼 삭제
x_test_datetime = x_test['datetime']        #추후 제출시 필요
x_train.drop(columns = 'datetime' , inplace = True)
x_test.drop(columns = 'datetime' , inplace = True)
y_train.drop(columns = '癤풼atetime' , inplace = True)
```

```python
# model 에서 제외할 feature columns 제외시키기!
x_train.drop(columns = 'month' , inplace = True)
x_train.drop(columns = 'dayofweek' , inplace = True)
```

```python
import sklearn.model_selection
print(dir(sklearn.model_selection))
```

```python
### 학습하고 평가하기
## 데이터 분리하기

# 데이터 분리를 위한 train_test_split 함수를 가져오기
from sklearn.model_selection import train_test_split

# 학습용과 검증용을 8:2로 분리하여, 각 4개의 변수에 저장하기
X_TRAIN, X_TEST, Y_TRAIN, Y_TEST = train_test_split(x_train, y_train, test_size = 0.2, random_state = 10)
print(Y_TEST.shape)

## count 칼럼 값을 예측하는 모델 만들기 (예측 모델 ==> XGBRegressor 활용)
# 첫번째 모델) 데이터 학습 및  하이퍼 파라미터 튜닝하기
from xgboost import XGBRegressor
model = XGBRegressor(n_estimators = 100, max_depth = 3, random_state = 10)
model.fit(X_TRAIN, Y_TRAIN)     # drop 하니까 제대로 시행
```

```python
model
```

```python
# 결과 예측하기
# 결과 예측하기 = y_test frame 만들기!
y_test_predicted = (pd.DataFrame(model                          #predict 함수로 추출된 결과는 dataframe 의 결과가 아님!
                                .predict(x_test))
                                .rename(columns = {0:'count'})  # 학습시킨 모델의 feature와 x_test의 feature 값이 달라서 오류***
                                )    
y_test_predicted.head()
```

```python
# y_test_predicted 의 'count' 칼럼값이 음수인데이터를 추출하여, 0으로 바꾸기 (대여량을 의미하는 지표이기 때문)
y_test_predicted[y_test_predicted['count'] < 0] = 0

# 학습이 완료된 모델을 통해 Y_TEST 값을 예측하기 : 평가지표 계산용
Y_TEST_PREDICTED = pd.DataFrame(model.predict(X_TEST)).rename(columns = {0:'count'})

# Y_TEST_PREDICTED 의 'count' 칼럼 값이 음수인 데이터를 추출하여 0으로 바꾸기
Y_TEST_PREDICTED[Y_TEST_PREDICTED['count'] < 0] = 0
```

```python
# 모델 평가하기
from sklearn.metrics import r2_score

print(r2_score(Y_TEST, Y_TEST_PREDICTED))
```

```python
# 두번째 모델) 데이터 학습 및  하이퍼 파라미터 튜닝하기

model = XGBRegressor(n_estimators = 200, max_depth = 5, random_state = 10)
model.fit(X_TRAIN, Y_TRAIN)

# 결과 예측하기
y_test_predicted = (pd.DataFrame(model                          #predict 함수로 추출된 결과는 dataframe 의 결과가 아님!
                                .predict(x_test))
                                .rename(columns = {0:'count'})  # 학습시킨 모델의 feature와 x_test의 feature 값이 달라서 오류***
                                )    

# y_test_predicted 의 'count' 칼럼값이 음수인데이터를 추출하여, 0으로 바꾸기 (대여량을 의미하는 지표이기 때문)
y_test_predicted[y_test_predicted['count'] < 0] = 0

# 학습이 완료된 모델을 통해 Y_TEST 값을 예측하기 : 평가지표 계산용
Y_TEST_PREDICTED = pd.DataFrame(model.predict(X_TEST)).rename(columns = {0:'count'})

# Y_TEST_PREDICTED 의 'count' 칼럼 값이 음수인 데이터를 추출하여 0으로 바꾸기
Y_TEST_PREDICTED[Y_TEST_PREDICTED['count'] < 0] = 0

# 모델 평가하기
from sklearn.metrics import r2_score

print(r2_score(Y_TEST, Y_TEST_PREDICTED))
```
